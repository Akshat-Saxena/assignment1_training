## Q.2 
- Plot a Sine Wave with color code with "black" color and style as dashdot, also mark "x" and "y" label and give a suitable title to it

```python
x=np.linspace(0,20,1000)
plt.plot(x,np.sin(x),color='black',linestyle='-.')
plt.xlabel('x',size=50)
plt.ylabel('y',size=50)
plt.title('sin curve',size=50)
```

## Q.4

-Plot Stacked Bar Graph for the above data.
    - Use proper colors and x and y label for the graph
    - Use legend positioned at upper right corner
    - Also mark xticks with an angle of 90 Degree

```python
store=['Delhi','Mumbai','Chennai','Kolkata','Hyderabad']
clothing=np.array([82,78,48,46,39])
equipment=np.array([48,31,51,29,18])
accessories=np.array([15,20,14,13,9])
var=[x for x, _ in enumerate(store)]
plt.bar(var,clothing,width=0.8,label="clothing",color='orange',bottom=equipment+accessories)
plt.bar(var,equipment,width=0.8,label="equipment",color='blue',bottom=accessories)
plt.bar(var,accessories,width=0.8,label="accessories",color='green')
plt.ylabel('quantity')
plt.xlabel('store')
plt.xticks(var,store)
plt.title('stores')
plt.grid(False)

plt.legend(loc='upper right')
plt.show()
```
